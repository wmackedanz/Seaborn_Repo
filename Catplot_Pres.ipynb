# CatPlots Showcase


"Catplot" is short for categorical plot

### Using catplot give access to all categorical plots

##### Including:
- Box plot
- violin plot
- boxenplot
- stripplot
- barplot
- countplot
- pointplot
- swarmplot

```python
import seaborn as sns
```

```python
cars = sns.load_dataset('mpg')
cars.dropna(inplace=True)

cars = cars[cars.cylinders.isin([4,6,8])]

cars['type'] = ['old' if x <= 76 else 'new' for x in cars.model_year]
```

```python
cars.head()
```

```python
cars.shape
```

```python
sns.set_style('white')
```

```python
#default catplot is a stripplot
sns.catplot(x='horsepower',data=cars);
```

```python
#use "kind =" to select what type of categorical plot you want
sns.catplot(x='horsepower',data=cars,kind = 'box');
```

```python
#Adding a Y
sns.catplot(x='origin',
            y = 'horsepower',
            data=cars,
            kind = 'box');
```

```python
#Adding other categories
sns.catplot(x='origin',
            y = 'horsepower',
            hue = 'cylinders',
            data=cars,
            kind = 'box');
```

## Catplot facet grid


```python
#By using the 'col' function, you can add figures based on categories
sns.catplot(x='cylinders',
            y = 'horsepower',
            data=cars,
            kind = 'box',
           col='origin');
```

```python
#Can go further with 'row'
sns.catplot(x='cylinders',
            y = 'horsepower',
            data=cars,
            kind = 'box',
           col='origin',
           row='type',
           row_order = ['new','old']); #can use [insert]_order to change the order of the row/columns in a list
```

## Styling
#### in general, keyword arguements will work as normal for intended plots

```python

sns.catplot(x='cylinders',
            y = 'horsepower',
            data=cars,
            kind = 'box',
           col='origin',
           row='type',
           palette = 'rocket',
           whis=1); 
```

```python
#Another Example
sns.catplot(x='cylinders',
            y = 'horsepower',
            data=cars,
            kind = 'strip',
           col='origin',
            size = 20
           ); 
```

## Facet Styling

#### catplot objects are saved as a FacetGrid type, so any styling you can apply to facet grids, you can apply to catplots

```python

plot = sns.catplot(x='cylinders',
            y = 'horsepower',
            data=cars,
            kind = 'box',
           col='origin');
```
